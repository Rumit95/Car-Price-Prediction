```python
x=df.loc[(df["make"]=="dodge") & (df["price"]>11000 )].index
y=df.loc[(df["make"]=="honda") & (df["price"]>11000 )].index
z=df.loc[(df["make"]=="isuzu") & (df["price"]>20000 )].index
w=df.loc[(df["make"]=="mitsubishi") & (df["price"]>13000 )].index
e=df.loc[(df["make"]=="plymouth") & (df["price"]>11000 )].index
q=df.loc[(df["make"]=="toyota") & (df["price"]>14000 )].index
```

```python
print(x,y,z,q,w,e)
l=[x,y,z,q,w,e]
```

```python
for i in l:
    df.drop(index=i,inplace=True)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df,x=df["price"],y=df["make"])
plt.show()
```

```python
dfob=df.select_dtypes(object)
dfnum=df.select_dtypes(exclude=[object])
df.head()
```

```python
dfob.head()
```

```python
dfnum.head()
```

```python
dfob.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
dfob[['make', 'fuel-type', 'body-style', 'drive-wheels', 'engine-location',
       'engine-type']]=oe.fit_transform(dfob[['make', 'fuel-type', 'body-style', 'drive-wheels', 'engine-location',
       'engine-type']])
```

```python
dfob.head()
```

```python
df2=pd.concat([dfob,dfnum],axis=1)
```

```python
df2.head()
```

```python
df2.corr()
```

```python
#drive-wheels, engine-location, engine-size, city-mpg ,highway-mpg
x=df2[["drive-wheels","engine-location","width","engine-size","horsepower","city-mpg","highway-mpg"]]
x
```

```python
y=df2.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
print("Accuracy of model ",round(r2_score(ytest,ypred)*100,2),"%")
```
